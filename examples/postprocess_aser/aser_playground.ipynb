```python
import pprint
import networkx as nx
from time import time
from random import sample
from copy import deepcopy
from tqdm import tqdm
```

```python
pp = pprint.PrettyPrinter(indent=2)

# print node information
def print_node_info(target):
    print(target)
    print('Frequency:', aser.nodes[target]['freq'])
    for info in aser.nodes[target]['info']:
        pp.pprint(eval(info))

# print neighbor information
def peek_neighbors(target, with_cooccur=True):
    print(target)
    print('------ successors -----')
    succ = {}
    for n in aser.successors(target):
        succ[n] = deepcopy(list(aser.edges[target, n].values())[0])
        
        if not with_cooccur: succ[n].pop('Co_Occurrence', _)
        print(n, succ[n])

    print('------ predecessors -----')
    pred = {}
    for n in aser.predecessors(target):
        pred[n] = deepcopy(list(aser.edges[n, target].values())[0])
        if not with_cooccur: pred[n].pop('Co_Occurrence', _)
        print(n, pred[n])
        
    return {'text': target, 'successors': succ, 'predecessors': pred}

```
